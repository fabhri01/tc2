## Simulación numérica ##

```python
# importo librerias
import numpy as np
import scipy.signal as sig
import matplotlib.pyplot as plt
#libreria de la catedra
from pytc2.sistemas_lineales import pzmap, GroupDelay, bodePlot, analyze_sys, pretty_print_SOS, tf2sos_analog
from pytc2.general import print_subtitle

#Funcion de aproximacion Maxima Planicidad
def MP_aprox(ws, alpha_max, alpha_min): #wp = 1
    
    e = np.sqrt(10**(alpha_max/10) - 1)
    
    orden = np.ceil(0.5*(np.log10((10**(alpha_min*0.1) - 1)/ e**2) / np.log10(ws)))
    
    z, p, k = sig.buttap(orden)
    num, den = sig.zpk2tf(z,p,k)
    num, den = sig.lp2lp(num, den, e**(-1/orden)) #Renormalizacion del filtro por usar wbutter
    
    z, p, k = sig.tf2zpk(num, den)
    
    return num, den, orden
```

```python
ws = 2
alpha_max = 1
alpha_min = 12

num, den, n = MP_aprox(ws, alpha_max, alpha_min)
T = sig.TransferFunction(num, den)

bodePlot(T, fig_id = 1)
pzmap(T, fig_id = 2)
GroupDelay(T, fig_id = 3)
```

```python
# Para verificar con lo hecho en papel

e = np.sqrt(10**(alpha_max/10) - 1)

a = 2*e**(-1/3)  
b = 2*e**(-2/3)
c = 1/e

num2 = [c]
den2 = [1, a, b, c]

z, p, k = sig.tf2zpk(num2, den2)

print("Numerador simulación numérica: ", num)
print("Denominador simulación numérica: ", den)
print("Numerador papel: ", num2)
print("Denominador papel: ", den2)
```

```python
# Otra manera de imprimir los gŕaficos con más detalle

all_sys = []
filter_names = []

all_sys.append(sig.TransferFunction(num, den))
label = 'Maxima-Planicidad' + '_orden-' + str(n) + '_ripple_' + str(alpha_max) + 'dB' # etiqueta
print_subtitle(label)

this_sos = tf2sos_analog(num, den) #factorizacion en sistemas de segundo orden (SOS)
pretty_print_SOS(this_sos, mode='omegayq')
filter_names.append(label)

plt.close('all')
analyze_sys(all_sys, filter_names)
```
