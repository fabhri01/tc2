## <u> Implementación circuital</u>

El esquema a seguir era el siguiente:

<center>



</center>

<center>

Figura 6: Conexionado circuital

</center>

Se tenía que colocar un filtro Antialias a la entrada, y también a la salida. Para esto se optó por diseñar un sencillo filtro pasabajos RC, cuya frecuencia de corte esté en 10KHz:

$$f_{c} = \dfrac{1}{2\pi RC}$$

Siendo que los componentes a mano eran una resistencia $1k2\Omega$ en serie con otra resistencia de $330\Omega$ y un capacitor de $10nF$, se obtuvo un valor cercano al esperado:

$$f_{c} = \dfrac{1}{2\pi*1530\Omega*10nF} = 10402,28Hz$$

Fue armado en un protoboard. La señal de salida del filtro antialias (de la entrada) se conectará al pin ADC_IN del LPC1769, y el pin ADC_OUT estará conectado a la entrada del filtro antialias de la salida. Las puntas del osciloscopio estarán en la salida de este último filtro y a partir de ahí se realizan las mediciones.

## <u> Mediciones </u> ##

#### <u> Ensayo del circuito </u> ####
El microcontrolador fue alimentado mediante su puerto USB, la magnitud de la señal senoidal de entrada fue de 2V pico a pico, con un nivel de continua de 1,5V para estar en el rango dinámico del ADC.

Se procedió con un barrido en frecuencia para medir amplitudes y desfasajes en puntos de las bandas de interés a sabiendas de que los filtros en la entrada y la salida tendrían un papel en la forma de éstas.

#### <u> Resultados </u> ####
En base a las mediciones realizadas se armaron tablas en Excel para computar los gráficos de las transferencias. Las mediciones de fase fueron canceladas debido a un error del software avisado por parte de los docentes en el momento. 

Posteriormente, cada filtro fue testeado por el analizador de audio.
