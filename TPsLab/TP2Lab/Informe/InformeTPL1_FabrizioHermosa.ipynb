## <u> Implementación circuital</u>

El esquema a seguir era el siguiente:

<center>



Figura 6: Conexionado circuital
</center>

Se tenía que colocar un filtro Antialias a la entrada, y también a la salida. Para esto se optó por diseñar un sencillo filtro pasabajos RC, cuya frecuencia de corte esté en 10KHz:

$$f_{c} = \dfrac{1}{2\pi RC}$$

Siendo que los componentes a mano eran una resistencia $1k2\Omega$ en serie con otra resistencia de $330\Omega$ y un capacitor de $10nF$, se obtuvo un valor cercano al esperado:

$$f_{c} = \dfrac{1}{2\pi*1530\Omega*10nF} = 10402,28Hz$$

Fue armado en un protoboard. La conexión de la señal de salida del filtro antialias se conectará al pin ADC_IN del LPC1769, y el pin ADC_OUT estará conectado a la entrada del filtro antialias de la salida. Las puntas del osciloscopio estarán en la salida de este último filtro y a partir de ahí se realizan las mediciones.

## <u> Mediciones </u> ##

#### <u> Ensayo del circuito </u> ####
